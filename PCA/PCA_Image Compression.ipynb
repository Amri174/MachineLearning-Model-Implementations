```python
import matplotlib.image as mplib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
#Read and reshapee Image
#Run th image with reduced reshaped chosen n pca components
#inverse transform
#display compressed image
```

```python
img = mplib.imread('r11.jpg')
```

```python
img.shape
```

```python
img_re = np.reshape(img,(148,148*3))
img_re.shape
```

```python
pca = PCA(40).fit(img_re)
img_trans = pca.transform(img_re)
```

```python
#inverse transform
img_Inv = pca.inverse_transform(img_trans)
img = np.reshape(img_Inv,(148,148,3))
```

```python
plt.axis('off') #X and Y axis not displayed|
```

```python
plt.imshow(img.astype('uint'))
plt.show()
```
