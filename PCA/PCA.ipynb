```python
D22=D11.drop(['Unnamed: 32'],axis=1)
X=D22
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 6)
pca.fit(X_scaled)
```

```python
print('Eigen values')
print(pca.explained_variance_)
print('Eigen vectors')
print(pca.components_)
```

```python
transformed_df = pd.DataFrame(pca.transform(X_scaled),columns=['PC1','PC2','PC3','PC4','PC5','PC6'])

transformed_df.to_csv("breastcancertransformed1.csv",index=False)
```

```python
#this is a plot sree plot to find the optimal number of components
#also known as elbow plot

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.xlabel('ideal columns')
plt.ylabel('size')
plt.plot(pca.explained_variance_,marker='o')
```
